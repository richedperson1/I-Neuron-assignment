# Task 1

```python
# Python program to check if the given number is a Disarium Number
def disarium(num):
    total = 0
    for power, i in enumerate(str(num)):
        total += int(i)**(power+1)
    if total == num:
        return "="
    else:
        return '!='


try:
    num = int(input("Enter number to check Disarium Number : "))
    result = disarium(num)
    print(f"Enter number {num} {result} Disarium Number")
except Exception as e:
    print(e)
```

# Task 2

```python
# Python program to print all disarium numbers between 1 to 100

try:
    
    for num in range(1,100+1):
        result = disarium(num)
        print(f"Enter number : {num} {result} Disarium Number")
except Exception as e:
    print(e)
```

# Task 3

```python
def happy(n):    
    r = s = 0;    
    while(n > 0):    
        r = n%10    
        s += r**2  
        n //= 10    
    return s  
try:

    n = int(input("Enter number to check happy number or not : "))    
    result = n;    

    while(result != 1 and result != 4):    
        result = happy(result)    
    if(result == 1):    
        print(f"{n} : Happy number")
    elif(result == 4):    
        print(f"{n} : Not happy number")  
        
except Exception as e:
    print(e)
```

# Task 5

```python
# Python program to determine whether the given number is a Harshad Number
def harshad_num(num):
    total = 0
    for n in str(num):
        total+=int(n)
    if num%total==0:
        return "is"
    else:
        return "is not"
    
try:
    num = int(input("Enter number to check Harshad number : "))
    print(f"The number {num} {harshad_num(num)} Harshad number ")

except Exception as e:
    print(e)
```
